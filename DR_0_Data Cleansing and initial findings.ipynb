# Group by month to see which months are most popular for postings

```python
df_month = df.groupby(df["Month"]).count()
df_month['jobpost'].plot(style='.-')
```

#job postings grouped at the month level. As can be seen in the plot, job postings peak in the months of March and June and are at a low in December and January
